```python
Train = CocaCola_Data.head(32)
Test = CocaCola_Data.tail(10)
Test
```

```python
import statsmodels.formula.api as smf 
```

#Linear Model

```python
linear_model = smf.ols('Sales~t',data=Train).fit()
pred_linear = pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2))
#np.sqrt(np.mean((Test['Footfalls']-np.array(pred_linear))**2))
rmse_linear
```

#Exponential

```python
Exp = smf.ols('log_st~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

#Quadratic 

```python
Quad = smf.ols('Sales~t+tsquare',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","tsquare"]]))
#pred_Quad = pd.Series(Exp.predict(pd.DataFrame(Test[["t","t_square"]))) # we hve to verify
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

#Additive Seasonality Quadratic 

```python
add_sea = smf.ols('Sales~month_2_Dec + month_2_Jun + month_2_Mar + month_2_Sep',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['month_2_Dec','month_2_Jun','month_2_Mar','month_2_Sep']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

#Additive Seasonality Quadratic 


```python
add_sea_Quad = smf.ols('Sales~t+tsquare+month_2_Dec+month_2_Jun+month_2_Mar+month_2_Sep',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['month_2_Dec','month_2_Jun','month_2_Mar','month_2_Sep','t','tsquare']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

##Multiplicative Seasonality

```python
Mul_sea = smf.ols('log_st~month_2_Dec+month_2_Jun+month_2_Mar+month_2_Sep',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

#Multiplicative Additive Seasonality 

```python
Mul_Add_sea = smf.ols('log_st~t+month_2_Dec+month_2_Jun+month_2_Mar+month_2_Sep',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea
```
